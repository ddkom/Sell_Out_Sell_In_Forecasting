```python
# find seasonality
from fhvisual.templates import LineChart
import matplotlib.pyplot as plt
import pandas as pd


class Seasonality_index:
    """ Builds a seasonality index using line chart.

    :param target_field: Column to run seasonality index against
    :dtype target_field: str

    :param linechart_kwargs: Formatting arguments to pass to LineChart
    :dtype linechart_kwargs: dict
    """

    def __init__(self, linechart_kwargs: dict = None) -> None:

        self._set_line_chart(linechart_kwargs)

    def _set_line_chart(self, linechart_kwargs: dict):
        if linechart_kwargs is None:
            linechart_kwargs = {'annot': False}

        self.line_chart = LineChart(**linechart_kwargs)

    def plot(self,
             target_field,
             df,
             cal_month_col,
             cal_yearmonth_col,
             linechart_kwargs: dict = {}) -> plt.Figure:
        seasonality_index_data = self._prep_data(target_field,
                                                 df,
                                                 cal_month_col,
                                                 cal_yearmonth_col,
                                                 covid=True)

        fig = self.line_chart.plot(seasonality_index_data, linechart_kwargs)

        return fig

    def _prep_data(self,
                   target_field,
                   df,
                   cal_month_col,
                   cal_yearmonth_col,
                   covid=True):
        """Processes the data for line graphing seasonality index for pre and post COVID or all years.
        Expected input is: calendar year and month, calendar month sales org, category-division, and target columns. 
        It is assumed that any subsetting of sales_org or category or division for example is done beforehand.
        """

        # All, Pre, and Post COVID frames
        # TODO adjust time ref so reading and reflecting the time in df
        # TODO include check for full years i.e. same # of rows per month col
        # TODO add functionality for title
        df['time_ref'] = 'Apr 2016 - Mar 2022'
        pre = df[df[cal_yearmonth_col] <= 202003][[col for col in df.columns
                                                  ]].reset_index(drop=True)
        pre['time_ref'] = 'Pre-COVID Apr 2016 - Mar 2020'
        post = df[df[cal_yearmonth_col] > 202003][[col for col in df.columns
                                                  ]].reset_index(drop=True)
        post['time_ref'] = 'Post-COVID Apr 2020 - Mar 2022'

        # For storing results
        results = []

        if covid:
            frames = [pre, post]
        else:
            frames = [df, pre, post]

        # Find seasonality index
        for data in frames:
            # Create a 12 month moving average beginning at the second month
            yearly_ma_series = round(data.iloc[1:, 4].rolling(window=12).mean(),
                                     3)
            # Insert starting at first month 6
            yearly_ma_series = yearly_ma_series.iloc[6:].reset_index(drop=True)
            data['ma12_5up'] = yearly_ma_series

            # Create a 12 month moving average beginning at the first month
            yearly_ma_series = round(data.iloc[:, 4].rolling(window=12).mean(),
                                     3)
            # insert starting at first month 6
            yearly_ma_series = yearly_ma_series.iloc[6:-1].reset_index(
                drop=True)
            data['ma12_6up'] = yearly_ma_series

            # Create an average of the 12 month moving averages to find the centered moving average
            data['centred_ma'] = round(
                data[['ma12_5up', 'ma12_6up']].mean(axis=1), 3)

            # Find the initial seasonal estimate
            data['initial_seasonal_estimate'] = round(
                data[target_field] / data['centred_ma'], 3)

            # Create a table for the average initial seasonal estimate and normalized seasonality by month
            ref_table = pd.DataFrame()
            # month cols
            ref_table[cal_month_col] = [
                '01', '02', '03', '04', '05', '06', '07', '08', '09', '10',
                '11', '12'
            ]
            ref_table['Month'] = [
                'Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep',
                'Oct', 'Nov', 'Dec'
            ]

            # Find the average initial seasonal estimate
            avg_of_initial_seasonal_estimates = round(
                data.groupby(cal_month_col).apply(
                    lambda x: x['initial_seasonal_estimate'].mean()), 3)
            avg_of_initial_seasonal_estimates = pd.DataFrame(
                avg_of_initial_seasonal_estimates).reset_index().rename(
                    columns={0: 'avg_of_initial_seasonal_estimates'})

            # merge
            ref_table = ref_table.merge(avg_of_initial_seasonal_estimates,
                                        on=cal_month_col,
                                        how='inner')

            # Find the normalized seasonality
            ref_table['normalized_seasonal_estimates'] = round(
                ref_table['avg_of_initial_seasonal_estimates'] /
                ref_table['avg_of_initial_seasonal_estimates'].sum() * 12,
                3) * 100

            # Save to frame
            period = data['time_ref'].unique()[0]
            y = (round(ref_table['normalized_seasonal_estimates'], 1).tolist())
            x = ref_table['Month'].tolist()

            # graphing
            graphing = pd.DataFrame(data={'Period': period, 'y': y, 'Month': x})
            # append results
            results.append(graphing)

        # combine the results
        combined_results = pd.concat(results)

        # save correct index order
        ordered_index = combined_results['Month'][:12].tolist()

        # pivot y into periods pre and post covid
        combined_results = combined_results.pivot_table(values=['y'],
                                                        index='Month',
                                                        columns=['Period'])

        # pivot causes alphabetic ordering of index, re-order index
        combined_results = combined_results.reindex(ordered_index)

        # pivot causes multilevel index, drop level
        combined_results = combined_results.droplevel(level=0, axis=1)

        return combined_results
```
